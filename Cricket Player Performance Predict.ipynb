```python
#iterative impute for BR column (ODIMatchResults_df)
df = pd.DataFrame(ODIMatchResults_df['BR'])

imputer = IterativeImputer(max_iter=10, random_state=0)
imputed_df = imputer.fit_transform(df)

# To Convert the imputed array back to a DataFrame
ODIMatchResults_imputed_df = pd.DataFrame(imputed_df, columns=df.columns)

# Replace the 'Target' column in the original DataFrame with imputed values
ODIMatchResults_df['BR'] = ODIMatchResults_imputed_df['BR']


print(ODIMatchResults_df)
```

```python
#iterative impute for Target column (ODIMatchTotals_df)
df = pd.DataFrame(ODIMatchTotals_df['Target'])

imputer = IterativeImputer(max_iter=10, random_state=0)
imputed_df = imputer.fit_transform(df)

# To Convert the imputed array back to a DataFrame
ODIMatchTotals_imputed_df = pd.DataFrame(imputed_df, columns=df.columns)

# Replace the 'Target' column in the original DataFrame with imputed values
ODIMatchTotals_df['Target'] = ODIMatchTotals_imputed_df['Target']


print(ODIMatchTotals_df)
```
