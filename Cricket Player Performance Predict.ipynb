```python
def describe_dataframes(dataframes):
    for df in dataframes:
        print("DataFrame Shape:")
        print(df.shape)
        print('*' * 100)
        print("DataFrame Info:")
        print(df.info())
        print('*' * 100)
        print("Missing Values Count:")
        print(df.isna().sum())
        print('*' * 100)
```

```python
# Split the 'Start Date' column into 'Day', 'Month', and 'Year'
import pandas as pd

# Assuming df is a list of DataFrames
for i in range(len(df)):
    if 'Start Date' in df[i].columns:
        df[i]['Start Date'] = pd.to_datetime(df[i]['Start Date'])
        df[i]['Year'] = df[i]['Start Date'].dt.year
        df[i]['Month'] = df[i]['Start Date'].dt.month
        df[i]['Day'] = df[i]['Start Date'].dt.day
        print("DataFrame", i)
        print(df[i].head())
        print("-------------------------")
    else:
        print("DataFrame", i, "does not have 'Start Date' column")
        print("-------------------------")
```

```python
#droping irrelevent columns that is: unnamed, Start Date, year
def drop_irrelevant_columns(df_list, columns_to_drop):
    modified_dfs = []  # List to store the modified DataFrames
    
    for i, df in enumerate(df_list):
        irrelevant_columns = [col for col in columns_to_drop if col in df.columns]
        
        if irrelevant_columns:
            df = df.drop(columns=irrelevant_columns, axis=1)
            modified_dfs.append(df)  # Store the modified DataFrame
            print("DataFrame", i, "after dropping irrelevant columns:")
            print(df.head())  # Printing the DataFrame after dropping columns
            print("-------------------------")
        else:
            print("DataFrame", i, "does not have any irrelevant columns")
            print("-------------------------")
    
    return modified_dfs

# Example usage:
columns_to_drop = ['Unnamed: 0', 'Start Date', 'Year']
drop_irrelevant_columns(df, columns_to_drop)
```

```python
#label encoder
from sklearn.preprocessing import LabelEncoder

numerics = ['int8', 'int16', 'int32', 'int64', 'float16', 'float32', 'float64']

def get_categorical_columns(df):
    cat_col = []  

    for col in df.columns:
        if df[col].dtype not in numerics:  # To Check if the column data type is non-numeric
            cat_col.append(col)

    return cat_col

def label_encode_categorical_columns(df_list):
    label = LabelEncoder()

    for i, df in enumerate(df_list):
        cat_cols = get_categorical_columns(df)
        
        for col in cat_cols:
            df_list[i][col] = label.fit_transform(df[col])

```

```python
#Data Preparation
# For Batting
#calculating batting average of each player
grand_df['Batting Average'] = grand_df['Bat1'] / grand_df['Inns']

sorted_df = grand_df[['Player', 'Batting Average']].sort_values(by='Batting Average', ascending=False)
print(sorted_df)
```

```python
# Calculating Strike Rate for each player
grand_df['Strike Rate'] = (grand_df['Runs_batsman'] / grand_df['BF']) * 100
print(grand_df[['Player', 'Strike Rate']])
```

```python
# Calculate centuries and half centuries for each player
grand_df['Centuries'] = grand_df['Runs_batsman'].apply(lambda x: 1 if x >= 100 else 0)
grand_df['HalfCenturies'] = grand_df['Runs_batsman'].apply(lambda x: 1 if (x >= 50 and x < 100) else 0)

# Group by 'Player' and sum the centuries and half centuries
player_stats = grand_df.groupby('Player')[['Centuries', 'HalfCenturies']].sum().reset_index()
print(player_stats)
```

```python
# Data Preparation
#For Bowling 
# Calculate Bowling Average for each player
grand_df['Bowling Average'] = grand_df['Runs_batsman'] / grand_df['Bat1']
print(grand_df[['Player', 'Bowling Average']])
```

```python
# Calculate Economy Rate for each player
grand_df['Economy Rate'] = (grand_df['Runs_batsman'] / (grand_df['BF'] / 6)).round(2)
print(grand_df[['Player', 'Economy Rate']])
```

```python
# Calculate Wickets Taken
grand_df['Wickets Taken'] = grand_df['Wkts_x'].sum()
print(grand_df[['Player', 'Wickets Taken']])
```

```python
# Calculate Bowling Strike Rate for each player
grand_df['Bowling Strike Rate'] = (grand_df['BF'] / grand_df['Wkts_x']).round(2)

# Display the Player and Bowling Strike Rate columns
print(grand_df[['Player', 'Bowling Strike Rate']])
```

```python
# Calculate the total Maiden Overs for each player
grand_df['Maiden Overs Total'] = grand_df['Mdns'].sum()
print(grand_df[['Player', 'Maiden Overs Total']])
```

```python
#Finding player performance score

# Calculate Batting Performance Score using Batting Average
grand_df['Batting Performance Score'] = grand_df['Batting Average'] * 10

# Calculate Bowling Performance Score using Bowling Average
grand_df['Bowling Performance Score'] = 100 / grand_df['Bowling Average']

# Calculate the Overall Performance Score as a combination of batting and bowling
grand_df['Overall Performance Score'] = (grand_df['Batting Performance Score'] + grand_df['Bowling Performance Score']) / 2

player_performance = grand_df[['Player', 'Overall Performance Score']]
print(player_performance)
```
